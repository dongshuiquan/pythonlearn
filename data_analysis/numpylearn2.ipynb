```python
import numpy as np
from scipy.stats import scoreatpercentile

data = np.loadtxt('mdrtb_2012.csv', delimiter=',', usecols=(1,), skiprows=1,unpack=True)

print 'Max method', data.max()
print 'Max function', np.max(data)

print 'Min method', data.min()
print 'Min function', np.min(data)

print 'Mean method', data.mean()
print 'Mean function', np.mean(data)

print 'Std method', data.std()
print 'Std function', np.std(data)

print 'Median', np.median(data)
print 'Score at percenttile 50', scoreatpercentile(data, 50)
```

```python
import numpy as np
A = np.mat("2 4 6; 4 ,2 6; 10, -4 18")
print A
inverse = np.linalg.inv(A)
print "inverse of A \n", inverse
print 'check\n', A * inverse
print 'error', A * inverse - np.eye(3)
```

```python
A = np.mat("1 -2 1; 0 2 -8; -4 5 9")
b = np.array([0, 8, -9])
print A,'\n',b
x = np.linalg.solve(A, b)
print 'solution', x
print 'check\n', np.dot(A, x)
```

```python
import numpy as np

A = np.mat("3, -2; 1, 0")
print "A\n", A

print "Eigevalues", np.linalg.eigvals(A)

eigenvalues, eigenvectors = np.linalg.eig(A)
print "First tuple of eig", eigenvalues
print "Second tuple of eig\n", eigenvectors

for i in range(len(eigenvalues)):
    print "Left", np.dot(A, eigenvectors[:,i])
    print "Right", eigenvalues[i] * eigenvectors[:,i]
    print
```

```python
import numpy as np
from matplotlib.pyplot import plot, show

cash = np.zeros(10000)
cash[0] = 1000
outcome = np.random.binomial(9, 0.5, size=len(cash))

for i in range(1, len(cash)):
    if outcome[i] < 5:
        cash[i] = cash[i - 1] - 1
    elif outcome[i] < 10:
        cash[i] = cash[i - 1] + 1
    else:
        raise AssertionError("Unexpected outcome " + outcome)
print outcome.min(), outcome.max()
plot(np.arange(len(cash)), cash)
show()
        
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000
normal_values = np.random.normal(size = N)
dummy, bins, dummy = plt.hist(normal_values, 100, normed=True, lw=1)
print dummy,"\n\n", bins
sigma = 1
mu = 0
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(bins - mu) ** 2) / (2 * sigma ** 2), lw=2)
plt.show()
```

```python
import numpy
import scipy
import matplotlib.pyplot as plt

ascent = scipy.misc.ascent()
random_mask = numpy.random.randint(0, 2, size=ascent.shape)
plt.subplot(221)
plt.title('Original')
plt.imshow(ascent)
plt.axis('off')

masked_array = numpy.ma.array(ascent, mask=random_mask)
print masked_array

plt.subplot(222)
plt.title('Masked')
plt.imshow(masked_array)
plt.axis('off')

plt.subplot(223)
plt.title('Log')
plt.imshow(numpy.log(ascent))
plt.axis('off')

plt.subplot(224)
plt.title('Log Masked')
plt.imshow(numpy.log(masked_array))
plt.axis('off')

plt.show()


```
